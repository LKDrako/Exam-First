// Importamos la librerias pandas y matplotlib las cuales se van a utilizar en el siguiente ejercisio

```python
import warnings
warnings.simplefilter('ignore', FutureWarning)
from pandas import *
import matplotlib.pyplot as plt
```

// Se hace el respectivo llamado de los datos que vamos a utilizar en la tabulacion, en este caso se hizo llamado desde los documentos en formato csv.

```python
csv='C:/Users/Unisimon/Documents/InteligenciaArtificial/Exam/data.csv'
data=read_csv(csv)
data
```

// Filtramos los datos del año 2019 de la tabla

```python
dfp = DataFrame()
dfp = data[data['Year'] == 2019]
dfp
```

// Extraemos los datos numericos de la tabla filtrada

```python
col1=dfp['Number of deaths due to tuberculosis, excluding HIV'].str.extract('(\d+(?:\.\d+)?)')
col1
```

// Extraemos los datos numericos de la segunda tabla, asi mismo los convertimos en Float ya que se presentaron en String, y asi mismo lo convertimos en Int

```python
colZ=dfp['Deaths due to tuberculosis among HIV-negative people (per 100 000 population)'].str.extract('(\d+(?:\.\d+)?)')
colX=colZ.astype(float)
col2=colX.astype(int)
col2
```

// Unimos Todas las columnas filtradas y limpiadas, asi mismo se reinicia el numero del index para conocer el valor numerico de la cantidad de paises

```python
df = DataFrame()
df['Country Name'] = dfp['Country']
df['Number of deaths due to tuberculosis, excluding HIV'] = col1
df['Deaths due to tuberculosis among HIV-negative people (per 100 000 population)'] = col2
df=df.reset_index()
df.drop(['index'], axis=1, inplace=True)
df.fillna(0, inplace=True)
df
```

// Selecionamos la Columna con la cual vamos a trabajar

# Datos del Problema

Muertes Por Tuberculosis en el 2019

```python
tbColumn = df['Deaths due to tuberculosis among HIV-negative people (per 100 000 population)']
tbColumn.sum()
```

El mayor número de muertes en un solo país son:

```python
tbColumn.max()
```

El menor número de muertes en un solo país son:

```python
tbColumn.min()
```

Calculo de la Media

```python
tbColumn.mean()
```

Calculo de la Mediana

```python
tbColumn.median()
```

Organizamos el listado por la columna de muertes por tuberculosis de menor a mayor para determinar los paises que tienen menos y mas muertes por tuberculosis

```python
df.sort_values('Deaths due to tuberculosis among HIV-negative people (per 100 000 population)')
```

Se calcula la tasa de mortalidad por cada 100.000 habitantes en cada pais durante el periodo del 2019

```python
populationColumn = df['Number of deaths due to tuberculosis, excluding HIV'].astype(float)
df['Muertes Por Tuberculosis (per 100,000)'] = (tbColumn * 100 / populationColumn).astype(float)
df.fillna(0, inplace=True)
df
```

Paises Con Tasa de Mortalidad del Mas Bajo al Mas Alto

```python
df.sort_values('Muertes Por Tuberculosis (per 100,000)')
```
